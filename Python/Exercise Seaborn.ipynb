# Seaborn Exercise

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanicDS = sns.load_dataset('titanic')
```

```python
titanicDS.head()
```

```python
titanicDS['sex'].count()
```

```python
sns.distplot(titanicDS['fare'],kde=False, bins=20)
```

```python
sns.jointplot(x='fare',y='age',data=titanicDS, color='blue')
```

```python
sns.swarmplot(x='class',y='age', data=titanicDS)
```

```python
sns.boxplot(x='class', y='age', data=titanicDS, palette='rainbow')
```

```python
sns.countplot(x='sex', data=titanicDS)
```

```python
titanicDSCor = titanicDS.corr()
```

```python
sns.heatmap(titanicDSCor, cmap='coolwarm')
plt.title('titanic_corr')
```

```python
sns.clustermap(titanicDSCor, cmap='coolwarm')
```

```python
g = sns.FacetGrid(data=titanicDS, col='sex')# , row='sex')
g.map(plt.hist, 'age')
```

```python
g = sns.FacetGrid(data=titanicDS, col='sex')# , row='sex')
g.map(sns.distplot, 'age')
```
